###Data Split

 From the heatmap, it is clear that `smoker_yes` has the strongest positive correlation with insurance charges, while `age and BMI ` also show moderate positive correlations. In contrast, features like children and `sex_male` exhibit weaker correlations with `charges`.

 Although these features have a low linear correlation with the target, I chose to keep them for both statistical and logical reasons.
Statistically, even features with weak direct correlations can contribute to model performance through non-linear relationships or interactions with other variables.

 Logically, factors such as the number of children and gender are known to influence health insurance costs in real-world scenarios, as insurers often consider family size and demographic information when determining premiums.

I dropped the region column because in the original dataset, the region column refers to US regions, which are not relevant for an African audience. Including this feature would not add meaningful predictive value for my users, since the categories do not correspond to any local or regional distinctions in Africa.

 Keeping the column could mean that my model might learn patterns that are only valid for the US population, leading to misleading or unfair predictions when applied elsewhere. By removing the region feature, the model would provide accurate and fair predictions for the target user base(Africa).

```python
X = numerical_data.drop(columns=['charges','region_northwest', 'region_southeast', 'region_southwest', 'region_northeast'])
Y = numerical_data['charges']
```

```python
print(X.head())
print(X.columns)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

###Standardization

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

###Linear Regression

```python
lr = LinearRegression()
lr.fit(X_train_scaled, y_train)
y_pred_lr = lr.predict(X_test_scaled)
```

```python
plt.figure(figsize=(8, 4))
plt.scatter(y_test, y_pred_lr, alpha=0.5)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--')
plt.xlabel('Actual Charges')
plt.ylabel('Predicted Charges')
plt.title('Linear Regression: Actual vs Predicted Charges')
plt.show()
```

###SGD Linear Regression

```python
from sklearn.linear_model import SGDRegressor
```

```python
sgd = SGDRegressor()
sgd.fit(X_train_scaled, y_train)
y_pred_sgd = sgd.predict(X_test_scaled)
```

```python
from sklearn.metrics import mean_squared_error
n_epochs = 100
sgd = SGDRegressor(max_iter=1, tol=None, random_state=42, learning_rate='invscaling', eta0=0.01, warm_start=True)

train_losses = []
test_losses = []

for epoch in range(n_epochs):
    sgd.partial_fit(X_train_scaled, y_train)
    y_train_pred = sgd.predict(X_train_scaled)
    y_test_pred = sgd.predict(X_test_scaled)
    train_loss = mean_squared_error(y_train, y_train_pred)
    test_loss = mean_squared_error(y_test, y_test_pred)
    train_losses.append(train_loss)
    test_losses.append(test_loss)

plt.plot(train_losses, label='Train Loss')
plt.plot(test_losses, label='Test Loss')
plt.xlabel('Epoch')
plt.ylabel('MSE Loss')
plt.title('SGDRegressor Loss Curve (Train vs Test)')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8, 4))
plt.scatter(y_test, y_pred_sgd, alpha=0.5)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--')
plt.xlabel('Actual Charges')
plt.ylabel('Predicted Charges')
plt.title('SDG: Actual vs Predicted Charges')
plt.show()
```

###Decision Tree Regression Model

```python
from sklearn.tree import DecisionTreeRegressor
dt = DecisionTreeRegressor(random_state=42)
dt.fit(X_train_scaled, y_train)
y_pred_dt = dt.predict(X_test_scaled)
```

```python
plt.figure(figsize=(8, 4))
plt.scatter(y_test, y_pred_dt, alpha=0.5)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--')
plt.xlabel('Actual Charges')
plt.ylabel('Predicted Charges')
plt.title('Decision Tree: Actual vs Predicted Charges')
plt.show()
```

###Random Forest Regression Model

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor()
rf.fit(X_train_scaled, y_train)
y_pred_rf = rf.predict(X_test_scaled)
```

```python
plt.figure(figsize=(8,4))
plt.scatter(y_test, y_pred_rf, alpha=0.5)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--')
plt.xlabel('Actual Charges')
plt.ylabel('Predicted Charges')
plt.title('Random Forest: Actual vs Predicted Charges')
plt.show()
```

###Model Evaluation

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

def print_metrics(name, y_true, y_pred):
    mse = mean_squared_error(y_true, y_pred)
    rmse = np.sqrt(mse)
    mae = mean_absolute_error(y_true, y_pred)
    r2 = r2_score(y_true, y_pred)
    print(f'{name}:')
    print('  RMSE:', rmse)
    print('  MAE:', mae)
    print('  R2:', r2)
    print()

print_metrics('Linear Regression', y_test, y_pred_lr)
print_metrics('Decision Tree', y_test, y_pred_dt)
print_metrics('Random Forest', y_test, y_pred_rf)
print_metrics('SGD', y_test, y_pred_sgd)
```

###My Best Model

After evaluating All of the regression modelsincluding Linear Regression, Decision Tree, Random Forest, and SGD, I selected the Random Forest model as the best performer for this task. This decision was based on the fact that Random Forest achieved the lowest RMSE and MAE, as well as the highest R score among all models tested. Specifically, it produced an RMSE of 4881.49, an MAE of 2805.11, and an R of 0.87, indicating that it not only predicts insurance charges more accurately but also explains a greater proportion of the variance in the data.
